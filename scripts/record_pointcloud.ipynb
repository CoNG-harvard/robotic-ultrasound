```python
import pyrealsense2 as rs
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a context object. This object owns the handles to all connected realsense devices
pipeline = rs.pipeline()
pipeline.start()
```

```python
frames = pipeline.wait_for_frames()
depth = frames.get_depth_frame()
dep = np.asanyarray(depth.get_data())
plt.figure()
plt.imshow(dep, cmap='gray')
np.save('/home/mht/catkin_ws/src/robotic-ultrasound/data/sample_depth_frame.npy', dep)
dep.shape
```

```python
frames = pipeline.wait_for_frames()
color = frames.get_color_frame()
dep = np.asanyarray(color.get_data())
plt.figure()
plt.imshow(dep, cmap='gray')
# np.save('/home/mht/catkin_ws/src/robotic-ultrasound/data/sample_color_frame.npy', dep)
dep.shape
```
