```python
import pickle as pkl
from matplotlib import pyplot as plt
import torch
import sys


sys.path.append("../../")


from scripts.USImageCV.models.unet import UNet
from scripts.USImageCV.utils.bunny import bunny_mask_predict, load_bunny_model,max_connected_area
```

```python

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = load_bunny_model('../../scripts/USImageCV/model_weights/vessel_model.pth',device)
```

```python
frames = data['frames']
wp = data['waypoints']
masks = []
coords = []
for f,w in zip(frames,wp):
    # plt.imshow(f)

    # plt.show()

    mask = bunny_mask_predict(model,f,device)

    # plt.imshow(mask)

    # plt.show()
    masks.append(mask)

    coords.append(w[:3])
```
